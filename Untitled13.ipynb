```python
x1 = np.linspace(-np.pi, np.pi, 100)
x2 = np.linspace(-np.pi, np.pi, 100)

y1 = np.cos(x1)
y2 = np.sin(x2)
```

```python
plt.plot(x1, y1, x2, y2)
plt.show()
```

```python
plt.figure(figsize = (10, 6), dpi = 80)
plt.plot(x1, y1)
plt.plot(x2, y2)
plt.show()
```

```python
plt.figure(figsize=(10,7), dpi = 90)
plt.plot(x1, y1,'r-.')
plt.plot(x2, y2,'b--')
plt.show()
```

```python
plt.figure(figsize = (10, 6),dpi = 90)
plt.plot(x1, y1, color= 'yellow')
plt.plot(x2,y2, color = 'green')
```

```python
plt.figure(figsize=(10,7), dpi = 90)
plt.plot(x1, y1, color= 'yellow',lw =2.6 )
plt.plot(x2,y2, color = 'green',lw = 2.6)
plt.show()
```

```python
plt.figure(figsize = (10,7), dpi = 90)
plt.plot(x1, y1, color='yellow',lw = 2.6,ls ='--')
plt.plot(x2, y2,color = 'green',lw = 2.6,ls='-.')
plt.show()
```

```python
plt.figure(figsize =(10,7),dpi=90)
plt.plot(x1, y2, color='yellow',lw =2.6,ls='--')
plt.plot(x2, y2, color='green',lw=2.6,ls='-.')
plt.xticks([-np.pi, -np.pi/2,0, np.pi/2, np.pi])
plt.yticks([1, 0, -1])
plt.show()
```

```python
plt.figure(figsize = (10, 7), dpi = 90)

plt.plot(x1, y1, color = 'yellow', lw = 2.6, ls = '--')
plt.plot(x2, y2, color = 'green', lw = 2.6, ls = '-.')

plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi], [r'$-\pi$', r'$-\pi/2$', r'$0$',r'$\pi/2$',r'$\pi$'])
plt.yticks([-1, 0, 1])

plt.show()
```
